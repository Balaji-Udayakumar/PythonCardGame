## Building a data structure containing the values of all cards in the deck

### Steps:

- created separate lists for each suit (hearts, diamonds, clubs, spades), each containing cards with values from 2 to 10.

- The individual suit lists are then combined into a single deck list by concatenating them together.

- Finally, used a for loop to iterate through the deck and print each card.

```python
# Create lists for each suit (hearts, diamonds, clubs, spades) containing cards with values from 2 to 10
hearts_list = [['hearts',2],['hearts',3],['hearts',4],['hearts',5],['hearts',6],['hearts',7],['hearts',8],['hearts',9],['hearts',10]]
diamonds_list = [['diamonds',2],['diamonds',3],['diamonds',4],['diamonds',5],['diamonds',6],['diamonds',7],['diamonds',8],['diamonds',9],['diamonds',10]]
clubs_list = [['clubs',2],['clubs',3],['clubs',4],['clubs',5],['clubs',6],['clubs',7],['clubs',8],['clubs',9],['clubs',10]]
spades_list = [['spades',2],['spades',3],['spades',4],['spades',5],['spades',6],['spades',7],['spades',8],['spades',9],['spades',10]]

# Combine all the suit lists to create a full deck
deck = hearts_list + diamonds_list + clubs_list + spades_list

# Loop through the deck and print each card
for card in deck:
    print(card)
```

## Function to distribute the deck

Steps:

- The random module is imported

- def distribute_deck(deck): defines a function named distribute_deck that takes one argument - deck

- Inside the function:
    - The random.sample function ensures that the selected cards are unique (no duplicates) and returns them as a list. These 18 cards are assigned to player1_hand, representing player 1's hand.
    -  In the next line used the filter function with a lambda function as a condition to include only those cards that are not already present in player1_hand. This ensures that player 2 receives the remaining cards.
    - Finally, the function returns both players' hands as a list

```python
import random

# Function to distribute the deck of cards equally to the two players
def distribute_deck(deck):

    # Randomly select 18 cards for player 1
    player1_hand = random.sample(deck,18)
    
    # Player 2 gets the remaining cards
    player2_hand = list(filter(lambda x: x not in player1_hand,deck))
    
    # Return both players hands
    return player1_hand,player2_hand
```

- Calculated and printed the sum of values of the card sets held by each player

```python

# invoking function distribute_deck and assigning the split deck to each of the players
player1_hand,player2_hand = distribute_deck(deck)

# Printing the cards of each player
print("Player 1 cards are :")
for player1_card in player1_hand:
    print(player1_card)
print("Player 2 cards are :")
for player2_card in player2_hand:
    print(player2_card)

# Printing the total numbers of cards held by each player
print(f"The number of cards dealt to player 1 is {len(player1_hand)} ")
print(f"The number of cards dealt to player 2 is {len(player2_hand)} ")

# Calculating the sum of values of the card sets held by each player
player1_card_sum = sum(player1_hand[1] for player1_hand in player1_hand)
player2_card_sum = sum(player2_hand[1] for player2_hand in player2_hand)

# Printing the sum of values of the card sets held by each player
print(f"The sum of values of the cards of player1 is {player1_card_sum} ")
print(f"The sum of values of the cards of player2 is {player2_card_sum} ")
```

## Play Game Steps

1. The function play_game takes two arguments, player1_hand and player2_hand, which represent the initial hands of Player 1 and Player 2, respectively.

2. Two variables, player1_exhausted and player2_exhausted, are used to track whether each player has run out of cards (exhausted) during the game in a tie. They are initially set to 0, indicating that both players have cards.

3. A variable rounds is used to keep track of the number of rounds played in the game.

4. The main game loop runs as long as the following conditions are met:

    - The number of rounds is less than 50.
    - Neither player has exhausted their cards during a tie (both player1_exhausted and player2_exhausted should be 0).
    - Both players have at least two cards in their hands.
5. Inside the loop, the code simulates a single round of the game:

    - It calculates the scores for both players in the current round based on the values of the first two cards in their hands.
    - It creates transfer sets (player1_transfer_set and player2_transfer_set) to hold the cards that will be transferred to the losing player in case of a tie.
    - It prints the cards played by both players and their scores for the current round.
    - It removes the first two cards from both players' hands.
    - It initializes the add_cards variable to 4, representing the number of cards to be added to the losing player's hand.
    
6. After calculating the scores for the current round, the code enters a nested loop to handle tie rounds:

    - If the scores of both players are equal and neither player has exhausted their cards, it's a tie.
    - The code continues drawing two cards from each player's hand and adding them to the tie round until one of the following conditions is met:
        - One player's hand is exhausted (either player1_exhausted or player2_exhausted becomes 1).
        - The scores are no longer tied.
    - During each iteration of the tie round, the code updates the round scores and transfer sets accordingly.
7. After exiting the tie round, the code checks whether any player's hand was exhausted during the tie round and, if so, returns the held cards to their original owners.

8. Further the code determines the winner of the round based on the scores. The losing player adds all the cards from the transfer sets (player1_transfer_set and player2_transfer_set) to their hand.

9. The code checks again if any player's hand has become exhausted after adding the transfer sets and updates the player1_exhausted and player2_exhausted flags accordingly.

10. The round counter is incremented, and the loop proceeds to the next round.

11. After exiting the loop, the code calculates the total scores for both players based on their remaining cards in hand.

12. Finally, the code determines and prints the result of the game, including which player wins or if it's a tie.

```python
# Function to play a single game
def play_game(player1_hand,player2_hand):
    
    # usage of exhausted flag to determine if players are out of cards
    player1_exhausted,player2_exhausted = 0,0
    
    # Variable to track rounds in a game
    rounds = 0
    
    # Main game loop with a limit of 50 rounds and the numbers of cards with each player should atleast be two
    # And if any player's hands is exhausted during a tie, the exhaust flags are turned to 1 and the loop exits as the game stops there
    while rounds < 50 and player1_exhausted == 0 and player2_exhausted == 0 and len(player1_hand) >= 2 and len(player2_hand) >= 2:
        print(f'Round {rounds+1}')
        
        # Calculate round scores for both players
        player1_round_score = sum(card[1] for card in player1_hand[:2])
        player2_round_score = sum(card[1] for card in player2_hand[:2])
        
        # Create transfer sets for both players to be able to transfer the cards to the losing player
        player1_transfer_set = player1_hand[:2]
        player2_transfer_set = player2_hand[:2]
        
        # Print cards played for each round
        print(f'Player 1 shows a {player1_transfer_set[0][1]} of {player1_transfer_set[0][0]} and a {player1_transfer_set[1][1]} of {player1_transfer_set[1][0]}, totalling {player1_round_score}')
        print(f'Player 2 shows a {player2_transfer_set[0][1]} of {player2_transfer_set[0][0]} and a {player2_transfer_set[1][1]} of {player2_transfer_set[1][0]}, totalling {player2_round_score}')
        
        # Remove first two cards from both players hands
        del player1_hand[:2]
        del player2_hand[:2]
        
        # Initialize variable to track additional cards to be added to the losing player for the round
        add_cards = 4

        # Handling rounds with equal scores (tie) and potential exhaustion of any of the player's hands in a tie round
        while player1_round_score == player2_round_score and player1_exhausted == 0 and player2_exhausted == 0:
            player1_round_set = []
            player2_round_set = []
            #Print that its a tie based on the scores
            print('Tie! Next cards are pulled')
            player1_exhausted = len(player1_hand) < 2
            player2_exhausted = len(player2_hand) < 2
            
            # Check for exhaustion for the 3 possible scenarios below
            if player1_exhausted and player2_exhausted:
                print('Both player 1 and player 2 cards are insufficient to play')
            elif player1_exhausted:
                print('Player 1 cards are insufficient to play')
            elif player2_exhausted:
                print('Player 2 cards are insufficient to play')
            # Continue the round with additional cards as they are available
            else:
                # Created a list called round_set to add the cards for a single play to the list in a given round during a tie
                player1_round_set.extend(player1_hand[:2])
                player2_round_set.extend(player2_hand[:2])
                
                # Calculating the play's scores for that round during a tie
                player1_round_score = sum(player1_round_set[1] for player1_round_set in player1_round_set)
                player2_round_score = sum(player2_round_set[1] for player2_round_set in player2_round_set)
                
                # Print the cards drawn for the play in a tie round
                print(f'Player 1 shows a {player1_round_set[0][1]} of {player1_round_set[0][0]} and a {player1_round_set[1][1]} of {player1_round_set[1][0]}, totalling {player1_round_score}')
                print(f'Player 2 shows a {player2_round_set[0][1]} of {player2_round_set[0][0]} and a {player2_round_set[1][1]} of {player2_round_set[1][0]}, totalling {player2_round_score}')                
                
                # Remove cards from players' hands
                del player1_hand[:2]
                del player2_hand[:2]
                
                # Adding back each player's cards drawn for a play to the transfer_set in a round during a tie
                player1_transfer_set.extend(player1_round_set)
                player2_transfer_set.extend(player2_round_set)
                
                # updating count of the additional cards the losing player must add
                add_cards = add_cards + 4
        
        # Condition to check if any one of the player's hands got exhausted during a tie
        # Incase if any one player is exhausted in a tie round the card sets drawn for that round are transferred to the original players
        if(player1_exhausted == 1 or player2_exhausted == 1):
            print('Due to insufficient cards to play, the held cards during the round are returned back to the original players')
            player1_hand.extend(player1_transfer_set)
            player2_hand.extend(player2_transfer_set)
        # Else determine the winner of the round and update players' hands using transfer_set list
        else:
            if player1_round_score < player2_round_score:
                print(f'Player 1 wins the round and Player 2 must add all {add_cards} cards to their card set.')
                player2_hand.extend(player1_transfer_set + player2_transfer_set)
            elif player1_round_score > player2_round_score:
                print(f'Player 2 wins the round and Player 1 must add all {add_cards} cards to their card set.')
                player1_hand.extend(player1_transfer_set + player2_transfer_set)
        
        # Condition to check if any player's hands is exhausted after adding the transfer sets to the losing player
        # Based on the player who got exhausted a print statement is executed
        if len(player1_hand) < 2:
            print('Player 1 cards are insufficient to play')
        elif len(player2_hand) < 2:
            print('Player 2 cards are insufficient to play')
        
        # Increment the round counter
        rounds += 1
    
    # Calculate the total scores for both players at the end of the game
    player1_total_score = sum(player1_hand[1] for player1_hand in player1_hand)
    player2_total_score = sum(player2_hand[1] for player2_hand in player2_hand)

    # Determine the player who won the game in the end
    if player1_total_score < player2_total_score:
        print('Player 1 wins the game !!!')
    elif player1_total_score > player2_total_score:
        print('Player 2 wins the game !!!')
    else:
        print('It''s a tie')
```

## Simulating running the game for 1000 times

The logic is the same as mentioned for the play_game function for which the steps have been listed in the markdown above. The print statements are now removed from the play_game function and the game is being run for 1000 times

Additional steps included in the code :

- Three lists (player_total_stats, player1_hand, player2_hand) are initialized to store game statistics, player hands, and the deck of cards.
- The distribute_deck(deck) function distributes the deck of cards randomly to two players: player1_hand and player2_hand.
- The play_game(player1_hand, player2_hand) function simulates a single game between the two players.
- Starting points for both players, player1_start_points and player2_start_points, are calculated by summing the values of the cards in their initial hands inside the play_game function.
- After each game, the total scores for both players are calculated and stored in the **player_total_stats** list as **[player1_start_points, player2_start_points, player1_total_score, player2_total_score]**.
- The code then simulates 1000 games in a loop, distributing the deck and playing the game for each iteration.
- Finally, it prints the entire player_total_stats list, which contains the game statistics for all simulated games.

```python
# Initialize empty lists to store game statistics and player hands
player_total_stats = []
player1_hand = []
player2_hand = []

# Function to distribute the deck of cards to two players
def distribute_deck(deck):
    # Randomly select 18 cards for player 1
    player1_hand = random.sample(deck,18)
    
    # Player 2 gets the remaining cards
    player2_hand = list(filter(lambda x: x not in player1_hand,deck))
    
    # Return both players hands
    return player1_hand,player2_hand

# Function to play a single game
def play_game(player1_hand,player2_hand):
    
    # Initialize variables to track per game statistics
    player_stats_per_game = []
    
    # usage of exhausted flag to determine if players are out of cards in a tie round and exit the loop if any of the players is out of cards
    player1_exhausted,player2_exhausted = 0,0
    
    # used to track if players are out of cards in a tie round
    # used to return back the cards to the original players if any one player is out of cards and the game ends there
    return_back_flag = 0
    
    # Variable to track rounds in a game
    rounds = 0
    
    # Calculate starting points for both players
    player1_start_points = sum(player1_hand[1] for player1_hand in player1_hand)
    player2_start_points = sum(player2_hand[1] for player2_hand in player2_hand)
    
    # Main game loop with a limit of 50 rounds and the numbers of cards with each player should atleast be two
    # And if any player's hands is exhausted during a tie, the return_back_flag is turned to 1 and the loop exits as the game stops there
    while rounds < 50 and len(player1_hand) >= 2 and len(player2_hand) >= 2 and return_back_flag == 0:
        
        # Calculate round scores for both players
        player1_round_score = sum(card[1] for card in player1_hand[:2])
        player2_round_score = sum(card[1] for card in player2_hand[:2])

        # Create transfer sets for both players to be able to transfer the cards to the losing player
        player1_transfer_set = player1_hand[:2]
        player2_transfer_set = player2_hand[:2]
        
        # Remove first two cards from both players hands
        del player1_hand[:2]
        del player2_hand[:2]
        
        # Initialize variable to track additional cards to be added to the losing player for the round
        add_cards = 4
        
        # Handling rounds with equal scores (tie) and potential exhaustion of any of the player's hands in a tie round
        while player1_round_score == player2_round_score and player1_exhausted == 0 and player2_exhausted == 0:
            player1_round_set = []
            player2_round_set = []
            player1_exhausted = len(player1_hand) < 2
            player2_exhausted = len(player2_hand) < 2
            
            # Check for exhaustion and transfer the card sets for that round to the original players
            if player1_exhausted and player2_exhausted:
                return_back_flag = 1
            elif player1_exhausted:
                return_back_flag = 1                
            elif player2_exhausted:
                return_back_flag = 1
            # Continue the round with additional cards as they are available
            else:
                # Created a list called round_set to add the cards for a single play to the list in a given round during a tie
                player1_round_set.extend(player1_hand[:2])
                player2_round_set.extend(player2_hand[:2])
                
                # Calculating the play's scores for that round during a tie
                player1_round_score = sum(player1_round_set[1] for player1_round_set in player1_round_set)
                player2_round_score = sum(player2_round_set[1] for player2_round_set in player2_round_set)
                
                # Remove cards from players' hands
                del player1_hand[:2]
                del player2_hand[:2]
                
                # Adding back each player's cards drawn for a play to the transfer_set in a round during a tie
                player1_transfer_set.extend(player1_round_set)
                player2_transfer_set.extend(player2_round_set)
                
                add_cards = add_cards + 4

        # Condition to check if any one of the player's hands got exhausted during a tie
        # Incase if any one player is exhausted the card in a tie round, sets drawn for that round are transferred to the original players
        if(return_back_flag == 1):
            player1_hand.extend(player1_transfer_set)
            player2_hand.extend(player2_transfer_set)
        # Determine the winner of the round and update players' hands using transfer_set list
        if player1_round_score < player2_round_score:
            player2_hand.extend(player1_transfer_set + player2_transfer_set)
        elif player1_round_score > player2_round_score:
            player1_hand.extend(player1_transfer_set + player2_transfer_set)

        # Increment the round counter
        rounds += 1
    
    # Calculate the total scores for both players at the end of the game
    player1_total_score = sum(player1_hand[1] for player1_hand in player1_hand)
    player2_total_score = sum(player2_hand[1] for player2_hand in player2_hand)
    
    # Store the game statistics in a list
    player_stats_per_game = [player1_start_points,player2_start_points,player1_total_score,player2_total_score]
    
    # Append the game statistics to the overall player_total_stats list upto 1000 games
    player_total_stats.append(player_stats_per_game)

# Simulate 1000 games
for i in range(1000):
    # Distribute the deck of cards to two players
    player1_hand, player2_hand = distribute_deck(deck)
    
    # Play the game with the given player hands
    play_game(player1_hand, player2_hand)

# Printing the entire player statistics for all simulated games
print(player_total_stats)
```

## Generating Descriptive Statistics for 1000 games

 - It calculates and displays descriptive statistics for both the winning and losing players in these games.
 - The code begins by importing the statistics module, which provides functions for performing various statistical calculations.
 - Two empty lists, winning_scores and losing_scores, are created. These lists will be used to store the scores of the winning and losing players in each game, respectively.
 - The `player_total_stats` list is iterated one game at a time, unpacking the values for both players' start points and total scores.
 - In each game, by determining the winning and losing player, the scores are respectively added to the `winning_scores` and `losing_scores` lists where when the total score of one player is lesser than the other player, then that score is added to the winning_scores list and the other score which is greater is added to the losing_scores list.
 - After iterating through all the games, various descriptive statistics are calculated for both the winning and losing groups.
 - Finally, the winning and losing player statistics are printed.
 
 ## Inference
 
 - From the mean scores of the winning and losing players we can see that the mean score of winning players is significantly less than the mean score of losing players, indicating that the winners are outperforming the losers in a bigger margin.





```python
import statistics as stats

# We have a list of player_total_stats for 1,000 games

# Create empty lists to store winning and losing scores
winning_scores = []
losing_scores = []

# Iterate through the player_total_stats for each game
for game_stats in player_total_stats:
    
    # Unpack the game statistics for both players
    player1_start_points, player2_start_points, player1_total_score, player2_total_score = game_stats

    # Determine the winning and losing player based on total scores
    if player1_total_score > player2_total_score:
        winning_scores.append(player2_total_score)
        losing_scores.append(player1_total_score)
    elif player2_total_score > player1_total_score:
        winning_scores.append(player1_total_score)
        losing_scores.append(player2_total_score)

# Calculate descriptive statistics for winning and losing scores
winning_mean = stats.mean(winning_scores)
winning_min = min(winning_scores)
winning_max = max(winning_scores)

losing_mean = stats.mean(losing_scores)
losing_min = min(losing_scores)
losing_max = max(losing_scores)

# Print the results
print("Winning Player Statistics:")
print(f"Mean Score: {winning_mean}")
print(f"Minimum Score: {winning_min}")
print(f"Maximum Score: {winning_max}")

print("\nLosing Player Statistics:")
print(f"Mean Score: {losing_mean}")
print(f"Minimum Score: {losing_min}")
print(f"Maximum Score: {losing_max}")
```

## Finding the highest total number of points a player started with, but still won the game

- Initialized a variable highest_starting_points to 0 which keeps track of the highest starting points achieved by a player who also won the game.

- The player_total_stats list is iterated one game at a time, unpacking the values for both players start points and total scores.

- In each game, the code determines which player won and which player lost based on their total scores. If player1_total_score is greater than player2_total_score, it means Player 2 won, and their starting points (player2_start_points) are recorded. Similarly if Player 1 wins, Player 1's starting points (player1_start_points) are recorded. If the game is a tie, an equal_flag is set to 1 to mark the tie.

- After determining the winning player's starting points and whether the game was a tie, the code checks if the winning player's starting points are higher than the current highest_starting_points. If the starting points are higher and the game wasn't a tie (indicated by equal_flag == 0), the highest_starting_points variable is updated with the new, higher value.

- Finally, the code prints out the highest starting points achieved by a player who also won the game.


```python
# A variable to keep track of the highest starting points achieved by a player who also won the game
highest_starting_points = 0

# Iterate through the game statistics for each game
for game_stats in player_total_stats:
    
    # Unpack the game statistics for both players
    player1_start_points, player2_start_points, player1_total_score, player2_total_score = game_stats
    
    # Introduced a flag to identify ties
    equal_flag = 0

    # Determine the winning player
    if player1_total_score > player2_total_score:
        winning_start_points = player2_start_points
    elif player2_total_score > player1_total_score:
        winning_start_points = player1_start_points
    else:
        # It's a tie, so the equal_flag is changed to 1
        equal_flag = 1

    # Check if the winning player's starting points are higher than the current maximum and also check if the equal flag is 0 to exclude ties
    if winning_start_points > highest_starting_points and equal_flag == 0:
        highest_starting_points = winning_start_points

# Print the result
print(f"The highest total number of points a player started with, but still won the game is: {highest_starting_points}")
```

## Calculating the percentage of games that had a winner who started the game with a higher total number of points compared to their opponent.

- In each game, we check if the winner had less total score than their opponent and also had a higher starting point. If this condition is met for either player, it increments the winning_games_with_higher_start_points counter to track such games.

- After iterating through all the games, we calculate the percentage of games where the winner started with more total points compared to their opponent.

- Finally, we print out the calculated percentage of games that had a winner who started the game with a higher total number of points compared to their opponent.

## Inference

The inference that can be drawn from this is that there are comparitively less winners who started the game with more points than their opponents and are able to win the game. This can be witnessed as shown in the percentage of winners who started the game with more points than their opponents being relatively less. This suggests starting with more points is correlated with a higher chance of losing the game.

```python
# Initialize a counter for games where the winner had more starting points than their opponent
winning_games_with_higher_start_points = 0

# Calculate the total number of games played
total_games_played = len(player_total_stats)

# Iterate through the game statistics for each game
for game_stats in player_total_stats:
    
    # Unpack the game statistics for both players
    player1_start_points, player2_start_points, player1_total_score, player2_total_score = game_stats
    
    # Check if the winner had both more total points and more starting points than their opponent
    if (player1_total_score > player2_total_score and player2_start_points > player1_start_points) or (player2_total_score > player1_total_score and player1_start_points > player2_start_points):
        # Increment the counter for games where the winner had more starting points
        winning_games_with_higher_start_points += 1

# Calculate the percentage of games where the winner had more starting points
percentage_games_winner_with_higher_start_points = (winning_games_with_higher_start_points/total_games_played) * 100

# Print the result
print(f"The percentage of games that had a winner who started the game with a higher \
total number of points compared to their \nopponent is : {percentage_games_winner_with_higher_start_points:.2f}%") 
```
